**Topik 1: Constraint Satisfaction Problems (CSP) â€“ Seperti Menyusun Puzzle Warna**

**Analogi:**

Bayangkan kamu sedang mewarnai peta provinsi di Indonesia. Aturannya sederhana:

***dua provinsi yang berbatasan tidak boleh memiliki warna yang sama.***

Kode Python di Google Colab:

Kode berikut menggunakan pustaka constraint untuk menyelesaikan masalah pewarnaan peta.

```python
#1: Install Paket python-constraint
# Menggunakan perintah !pip install untuk menginstal paket python-constraint untuk masalah Constraint Satisfaction Problem (CSP).
!pip install python-constraint # Install the python-constraint package

#2: Mengimpor Modul Problem dari constraint
from constraint import Problem
# Mengimpor kelas Problem dari pustaka constraint, yang digunakan untuk mendefinisikan dan menyelesaikan CSP.

#3: Membuat Objek CSP
problem = Problem()
#Membuat objek Problem() yang merepresentasikan permasalahan CSP.

#4-5: Daftar Wilayah dan Hubungan Batasan
wilayah = ["A", "B", "C", "D"]
batasan = [("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")]
# Mendefinisikan 4 wilayah (A, B, C, D) dan batasannya, di mana dua wilayah bertetangga tidak boleh memiliki warna yang sama.

#6: Daftar Warna
warna = ["Merah", "Biru", "Hijau"]
# Menentukan warna yang tersedia untuk digunakan dalam pewarnaan wilayah.

#7-8: Tambahkan variabel CSP dengan domain warna
for w in wilayah:
    problem.addVariable(w, warna)
# Menambahkan variabel ke dalam CSP, di mana setiap wilayah memiliki domain warna yang bisa dipilih (Merah, Biru, atau Hijau).


#9: Tambahkan constraint: Wilayah yang berbatasan tidak boleh memiliki warna yang sama
for w1, w2 in batasan:
    problem.addConstraint(lambda c1, c2: c1 != c2, (w1, w2))
#adalah fungsi anonim (lambda function) yang menerima dua parameter (c1 dan c2),
#kemudian mengembalikan True jika c1 tidak sama dengan c2, dan False jika mereka sama.
# Dalam konteks ini, c1 dan c2 mewakili warna dua wilayah yang berbatasan.
# menunjukkan bahwa fungsi lambda tersebut diterapkan pada pasangan wilayah (w1, w2) yang berbatasan.

#10-11: Mencari dan Menampilkan Solusi
solusi = problem.getSolutions() #Mencari semua solusi yang memenuhi constraint.
print(f"Jumlah solusi yang ditemukan: {len(solusi)}") #Menampilkan jumlah solusi yang ditemukan.
for sol in solusi[:5]:  #Menampilkan hingga 5 solusi pertama sebagai contoh.
    print(sol)
```

**Instruksi Eksperimen di Colab:**

1. Tambahkan lebih banyak wilayah ke dalam daftar wilayah dan batasan.

2. Tambahkan warna baru, misalnya "Kuning", ke dalam daftar warna.

3. Ubah batasan, misalnya buat wilayah "A" berbatasan dengan semua wilayah lain.

4. Bandingkan jumlah solusi sebelum dan setelah perubahan.

**Berikut adalah versi modifikasi kode yang mengikuti instruksi eksperimen di Google Colab:**

1. Menambahkan lebih banyak wilayah: Tambah "E" dan "F".

2. Menambahkan lebih banyak batasan: "A" sekarang berbatasan dengan semua wilayah lainnya.

3. Menambahkan warna baru: Tambah "Kuning" ke daftar warna.

4. Bandingkan jumlah solusi sebelum dan sesudah perubahan.

```python
from constraint import Problem

# Fungsi untuk menghitung jumlah solusi dari suatu CSP
def hitung_solusi(wilayah, batasan, warna):
    problem = Problem()

    # Tambahkan variabel dengan domain warna
    for w in wilayah:
        problem.addVariable(w, warna)

    # Tambahkan constraint: Wilayah yang berbatasan tidak boleh memiliki warna yang sama
    for w1, w2 in batasan:
        problem.addConstraint(lambda c1, c2: c1 != c2, (w1, w2))

    # Hitung solusi
    solusi = problem.getSolutions()
    return solusi

# **Eksperimen 1: Data Awal**
wilayah_awal = ["A", "B", "C", "D"]
batasan_awal = [("A", "B"), ("A", "C"), ("B", "D"), ("C", "D")]
warna_awal = ["Merah", "Biru", "Hijau"]

solusi_awal = hitung_solusi(wilayah_awal, batasan_awal, warna_awal)
print(f"Jumlah solusi sebelum perubahan: {len(solusi_awal)}")

# **Eksperimen 2: Data yang Dimodifikasi**
wilayah_baru = ["A", "B", "C", "D", "E", "F"]
batasan_baru = [
    ("A", "B"), ("A", "C"), ("A", "D"), ("A", "E"), ("A", "F"),
    ("B", "D"), ("C", "D"), ("D", "E"), ("E", "F")
]
warna_baru = ["Merah", "Biru", "Hijau", "Kuning"]  # Warna baru ditambahkan

solusi_baru = hitung_solusi(wilayah_baru, batasan_baru, warna_baru)
print(f"Jumlah solusi setelah perubahan: {len(solusi_baru)}")

# **Bandingkan beberapa solusi sebelum dan sesudah perubahan**
print("\nContoh solusi sebelum perubahan:")
for sol in solusi_awal[:5]:
    print(sol)

print("\nContoh solusi setelah perubahan:")
for sol in solusi_baru[:5]:
    print(sol)
```

**Topik 2: Rule-Based AI â€“ Seperti Dokter Virtual yang Membantu Diagnosis**

**Analogi:**

Bayangkan kamu adalah dokter virtual. Setiap pasien datang dengan keluhan, dan kamu memiliki daftar aturan untuk mencocokkan gejala dengan penyakit.

**Kode Python di Google Colab:**

Kode ini menggunakan aturan berbasis if-else untuk diagnosis penyakit.

```python
def diagnose_penyakit(gejala):
    """
    Fungsi untuk mendiagnosis penyakit berdasarkan gejala yang diberikan.
    Gejala dicocokkan dengan aturan yang telah ditentukan sebelumnya.
    """

    # Dictionary aturan yang berisi pasangan gejala dan penyakit terkait
    aturan = {
        ("demam", "batuk"): "Flu",  # Jika pasien mengalami demam dan batuk, kemungkinan Flu
        ("demam", "sesak napas"): "COVID-19",  # Jika pasien mengalami demam dan sesak napas, kemungkinan COVID-19
        ("sakit kepala", "mual"): "Migren"  # Jika pasien mengalami sakit kepala dan mual, kemungkinan Migren
    }

    # Loop untuk mencocokkan gejala yang diberikan dengan aturan yang ada
    for key in aturan:  # Iterasi setiap pasangan gejala dalam dictionary aturan
        if set(gejala).issubset(set(key)):  # Memeriksa apakah gejala pasien merupakan subset dari aturan
            return f"Diagnosis: Kemungkinan Anda terkena {aturan[key]}"  # Mengembalikan hasil diagnosis

    return "Gejala tidak dikenali. Silakan konsultasi ke dokter."  # Jika gejala tidak ditemukan dalam aturan

# Contoh penggunaan
gejala_pasien = ["demam", "batuk"]  # Daftar gejala yang dialami pasien
print(diagnose_penyakit(gejala_pasien))  # Memanggil fungsi dan mencetak hasil diagnosis
```

**Instruksi Eksperimen di Colab:**

1. Tambahkan lebih banyak aturan dalam dictionary aturan, misalnya untuk penyakit lain seperti tifus.

2. Tambahkan fitur input pengguna, agar Anda bisa memasukkan gejala secara manual.

3. Buat sistem lebih fleksibel, misalnya dengan menggunakan persentase kecocokan dibanding aturan ketat.

4. Bandingkan hasil diagnosis sebelum dan sesudah perubahan aturan.

```python
def diagnose_penyakit(gejala):
    """
    Fungsi untuk mendiagnosis penyakit berdasarkan gejala yang diberikan.
    Gejala dicocokkan dengan aturan yang telah ditentukan sebelumnya.
    """

    # Dictionary aturan yang berisi pasangan gejala dan penyakit terkait
    aturan = {
        ("demam", "batuk"): "Flu",  # Jika pasien mengalami demam dan batuk, kemungkinan Flu
        ("demam", "sesak napas"): "COVID-19",  # Jika pasien mengalami demam dan sesak napas, kemungkinan COVID-19
        ("sakit kepala", "mual"): "Migren",  # Jika pasien mengalami sakit kepala dan mual, kemungkinan Migren
        ("demam", "mual", "sakit perut"): "Tifus",  # Menambahkan aturan untuk penyakit Tifus
        ("demam", "menggigil", "nyeri sendi"): "Dengue"  # Menambahkan aturan untuk Demam Berdarah
    }

    # Menentukan tingkat kecocokan gejala dengan aturan yang ada
    best_match = None
    best_score = 0

    for key in aturan:  # Iterasi setiap pasangan gejala dalam dictionary aturan
        match_score = len(set(gejala) & set(key)) / len(set(key))  # Hitung persentase kecocokan
        if match_score > best_score:  # Simpan aturan dengan kecocokan tertinggi
            best_match = aturan[key]
            best_score = match_score

    if best_score > 0:  # Jika ada kecocokan, berikan hasil diagnosis
        return f"Diagnosis: Kemungkinan Anda terkena {best_match} dengan kecocokan {best_score:.0%}"

    return "Gejala tidak dikenali. Silakan konsultasi ke dokter."  # Jika gejala tidak ditemukan dalam aturan

# Input dari pengguna
gejala_pasien = input("Masukkan gejala yang Anda alami (pisahkan dengan koma): ").lower().split(", ")  # Meminta input gejala
print(diagnose_penyakit(gejala_pasien))  # Memanggil fungsi dan mencetak hasil diagnosis
```

**Topik 3: AI Planning â€“ Seperti Robot Kurir Mengantar Paket**

**Analogi:**
Bayangkan kamu memiliki robot kurir yang bisa mengantarkan paket ke berbagai lokasi di kota.


**Kode Python di Google Colab:**

Kode ini menggunakan pustaka networkx untuk merencanakan rute terpendek bagi robot.

```python
import networkx as nx  # Import library NetworkX untuk pemodelan graf

# Buat graf kota dan jalurnya
kota = nx.Graph()  # Inisialisasi graf tidak berarah
kota.add_edges_from([
    ("Gudang", "A", {"jarak": 5}),  # Tambahkan jalur dari Gudang ke A dengan jarak 5
    ("A", "B", {"jarak": 7}),  # Tambahkan jalur dari A ke B dengan jarak 7
    ("A", "C", {"jarak": 3}),  # Tambahkan jalur dari A ke C dengan jarak 3
    ("B", "D", {"jarak": 4}),  # Tambahkan jalur dari B ke D dengan jarak 4
    ("C", "D", {"jarak": 2}),  # Tambahkan jalur dari C ke D dengan jarak 2
    ("D", "Tujuan", {"jarak": 6})  # Tambahkan jalur dari D ke Tujuan dengan jarak 6
])

# Cari rute terpendek dari Gudang ke Tujuan berdasarkan bobot jarak
rute = nx.shortest_path(kota, source="Gudang", target="Tujuan", weight="jarak")  # Menggunakan algoritma Dijkstra secara default
print("Rute tercepat:", " -> ".join(rute))  # Cetak rute dengan format yang mudah dibaca
```

Instruksi Eksperimen di Colab:

1. Tambahkan lebih banyak lokasi dan jalur di dalam kota.add_edges_from().
2. Ubah bobot jarak, misalnya buat beberapa jalur lebih pendek atau lebih panjang.
3. Tambahkan beberapa robot, lalu cari rute optimal untuk masing-masing robot.
4. Bandingkan hasil rute sebelum dan sesudah perubahan untuk melihat bagaimana AI Planning bekerja.

```python
import networkx as nx  # Import library NetworkX untuk pemodelan graf

# Buat graf kota dan jalurnya dengan lebih banyak lokasi dan jalur
kota = nx.Graph()  # Inisialisasi graf tidak berarah
kota.add_edges_from([
    ("Gudang", "A", {"jarak": 5}),  # Tambahkan jalur dari Gudang ke A dengan jarak 5
    ("A", "B", {"jarak": 7}),  # Tambahkan jalur dari A ke B dengan jarak 7
    ("A", "C", {"jarak": 3}),  # Tambahkan jalur dari A ke C dengan jarak 3
    ("B", "D", {"jarak": 4}),  # Tambahkan jalur dari B ke D dengan jarak 4
    ("C", "D", {"jarak": 2}),  # Tambahkan jalur dari C ke D dengan jarak 2
    ("D", "Tujuan", {"jarak": 6}),  # Tambahkan jalur dari D ke Tujuan dengan jarak 6
    # Lokasi tambahan
    ("Gudang", "X", {"jarak": 4}),
    ("X", "C", {"jarak": 6}),
    ("X", "B", {"jarak": 5}),
    ("C", "Tujuan", {"jarak": 4})
])

# Cari rute terpendek sebelum perubahan bobot
rute_awal = nx.shortest_path(kota, source="Gudang", target="Tujuan", weight="jarak")
print("Rute tercepat sebelum perubahan:", " -> ".join(rute_awal))

# Perubahan bobot jarak
kota["A"]["B"]["jarak"] = 5  # Memendekkan jalur A ke B
kota["C"]["D"]["jarak"] = 4  # Memperpanjang jalur C ke D

# Cari rute terpendek setelah perubahan bobot
rute_setelah = nx.shortest_path(kota, source="Gudang", target="Tujuan", weight="jarak")
print("Rute tercepat setelah perubahan:", " -> ".join(rute_setelah))

# Menambahkan beberapa robot dengan posisi awal berbeda
robot_posisi = {"Robot1": "Gudang", "Robot2": "X"}

# Cari rute optimal untuk masing-masing robot
for robot, start in robot_posisi.items():
    rute_robot = nx.shortest_path(kota, source=start, target="Tujuan", weight="jarak")
    print(f"{robot} rute tercepat: {' -> '.join(rute_robot)}")
```

```python
import networkx as nx  # Import library NetworkX untuk pemodelan graf
import matplotlib.pyplot as plt  # Import matplotlib untuk visualisasi graf

# Buat graf kota dan jalurnya
kota = nx.Graph()  # Inisialisasi graf tidak berarah
kota.add_edges_from([
    ("Gudang", "A", {"jarak": 5}),  # Tambahkan jalur dari Gudang ke A dengan jarak 5
    ("A", "B", {"jarak": 7}),  # Tambahkan jalur dari A ke B dengan jarak 7
    ("A", "C", {"jarak": 3}),  # Tambahkan jalur dari A ke C dengan jarak 3
    ("B", "D", {"jarak": 4}),  # Tambahkan jalur dari B ke D dengan jarak 4
    ("C", "D", {"jarak": 2}),  # Tambahkan jalur dari C ke D dengan jarak 2
    ("D", "Tujuan", {"jarak": 6}),  # Tambahkan jalur dari D ke Tujuan dengan jarak 6
    ("Gudang", "X", {"jarak": 4}),  # Tambahkan jalur baru Gudang ke X
    ("X", "C", {"jarak": 5}),  # Tambahkan jalur baru X ke C
    ("X", "B", {"jarak": 6}),  # Tambahkan jalur baru X ke B
    ("C", "Tujuan", {"jarak": 7})  # Tambahkan jalur baru C ke Tujuan
])

# Cari rute terpendek dari Gudang ke Tujuan berdasarkan bobot jarak
rute = nx.shortest_path(kota, source="Gudang", target="Tujuan", weight="jarak")  # Menggunakan algoritma Dijkstra secara default
print("Rute tercepat:", " -> ".join(rute))  # Cetak rute dengan format yang mudah dibaca

# Visualisasi graf
plt.figure(figsize=(8,6))  # Tentukan ukuran figure
pos = nx.spring_layout(kota)  # Tentukan layout graf
nx.draw(kota, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=10)  # Gambar graf
edge_labels = {(u, v): d["jarak"] for u, v, d in kota.edges(data=True)}  # Tambahkan label bobot pada jalur
nx.draw_networkx_edge_labels(kota, pos, edge_labels=edge_labels)
plt.title("Graf Kota dan Jalurnya")  # Tambahkan judul
plt.show()  # Tampilkan visualisasi
```

**Machine Learning** (*Linear Regression*)

```python
# Linear Regression
from sklearn.linear_model import LinearRegression
import numpy as np

# Data sample
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

model = LinearRegression()
model.fit(X, y)

prediksi = model.predict(X)
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
print("Prediksi:", prediksi)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data sample
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 3, 5, 7, 11])

# Membuat dan melatih model Linear Regression
model = LinearRegression()
model.fit(X, y)

# Melakukan prediksi menggunakan model yang telah dilatih
prediksi = model.predict(X)

# Menampilkan Koefisien, Intercept, dan Prediksi
print("Koefisien:", model.coef_)
print("Intercept:", model.intercept_)
print("Prediksi:", prediksi)

# Plot data asli
plt.scatter(X, y, color='blue', label='Data Asli')

# Plot garis regresi
plt.plot(X, prediksi, color='red', label='Garis Regresi')

# Menambahkan judul dan label
plt.title('Linear Regression')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title("Regression analysis")
plt.ylabel("X")
plt.xlabel("y")

# Menampilkan plot
plt.show()
```

**Machine Learning (*K-Means Clustering*)**

```python
# K-Means Clustering
from sklearn.cluster import KMeans

# Data sample
data = np.array([[1, 2], [1, 4], [1, 0],
                 [10, 2], [10, 4], [10, 0]])

kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

print("Cluster Centers:", kmeans.cluster_centers_)
print("Labels:", kmeans.labels_)
```

**Machine Learning (*K-Means Clustering dengan Plot*)**

---



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Data sample
data = np.array([[1, 2], [1, 4], [1, 0],
                 [10, 2], [10, 4], [10, 0]])

# Membuat dan melatih model K-Means
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Mendapatkan hasil clustering
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Menampilkan Cluster Centers dan Labels
print("Cluster Centers:", cluster_centers)
print("Labels:", labels)

# Plot data asli dan hasil clustering
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', label='Data Points')
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', s=200, alpha=0.75, label='Cluster Centers')

# Menambahkan judul dan label
plt.title('K-Means Clustering')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

# Menampilkan plot
plt.show()
```
