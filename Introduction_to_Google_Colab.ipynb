**Struktur Google Colab**

Google Colab memiliki struktur utama sebagai berikut:
1. Notebook - Dokumen tempat pengguna menulis dan menjalankan kode.
2. Sel Kode (Code Cell) - Tempat untuk menulis skrip Python.
3. Sel Teks (Text Cell) - Untuk menambahkan penjelasan dalam Markdown.
4. Panel Navigasi (Sidebar) - Untuk mengakses file dan pengaturan.
5. Runtime - Mengatur eksekusi kode dengan CPU/GPU/TPU.

#Topik 2: Magic Commands di Google Colab – Benarkah Membantu?

🔹 Penjelasan:
Google Colab memiliki fitur Magic Commands, yaitu perintah khusus yang memudahkan kita dalam menulis kode dan menganalisis data dengan lebih efisien.

Beberapa Magic Commands yang sering digunakan:

!ls → Melihat daftar file dalam direktori kerja.

%timeit some_code → Mengukur waktu eksekusi suatu kode.

%matplotlib inline → Menampilkan grafik langsung di dalam notebook.
